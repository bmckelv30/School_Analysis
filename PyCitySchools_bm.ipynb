```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Gather info about dataframe
# list (school_data_complete )
# school_data_complete.count()
school_data_complete.head()
# Array of school names
school_data_complete["school_name"].unique()
```

```python
# Calculate the total number of schools
total_sch = school_data_complete["school_name"].value_counts()
# total_sch.count()
```

```python
# Calculate the total number of students
total_students = school_data_complete["student_name"].count()
# total_students
```

```python
# Calculate the total budget
total_budget = school_data_complete["budget"].unique()
# total_budget.sum()
```

```python
# Calculate the average math score
avg_math = school_data_complete["math_score"].mean()
# avg_math
```

```python
# Calculate the average reading score
avg_read = school_data_complete["reading_score"].mean()
# avg_read
```

```python
# Calculate the passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_avg = (avg_math + avg_read)/2
# overall_avg
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
pass_math = school_data_complete.loc[school_data_complete["math_score"] > 69]
m = pass_math["math_score"] .count()
math_passing = m/total_students * 100
# math_passing
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
pass_read = school_data_complete.loc[school_data_complete["reading_score"] > 69]
r = pass_read["reading_score"] .count()
read_passing = r/total_students * 100
# read_passing
```

```python
# Create a dataframe to hold the above results
summary_table = pd.DataFrame({"Total Schools": [total_sch.count()],
                              "Total Students": [total_students],
                              "Total Budget": [total_budget.sum()],
                              "Average Math Score": [avg_math],
                              "Average Reading Score": [avg_read],
                              "% Passing Math": [math_passing],
                              "% Passing Reading": [read_passing],
                              "% Overall Passing Rate": [overall_avg],
                             })
# Optional: give the displayed data cleaner formatting
    # Total budget: Format with dollar sign and comma
summary_table["Total Budget"] = summary_table["Total Budget"].astype(float).map("${:,.2f}".format)
    # Total students: Format with comma
summary_table["Total Students"] = summary_table["Total Students"].astype(float).map("{:,}".format)
    # Average Math and Reading; Overall Rate; Math and Reading passing: Format to 6 decimal places
summary_table["Average Math Score"] = summary_table["Average Math Score"].map("{:.6f}".format)
summary_table["Average Reading Score"] = summary_table["Average Reading Score"].map("{:.6f}".format)
summary_table["% Passing Math"] = summary_table["% Passing Math"].map("{:.6f}".format)
summary_table["% Passing Reading"] = summary_table["% Passing Reading"].map("{:.6f}".format)
summary_table["% Overall Passing Rate"] = summary_table["% Overall Passing Rate"].map("{:.6f}".format)
summary_table
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Create DataFrame of the Schools to calculate each school's Total Students (size), Budget, and Type
schools = pd.read_csv(school_data_to_load)
schools = schools.rename(index=str, columns={"school_name":"School Name", 
                                                 "type":"Type",
                                                 "size":"Total Students",
                                                 "budget": "Total Budget"
                                                })
schools.drop ("School ID", axis=1, inplace=True)
# schools
```

```python
# Calculate each school's average Math and Reading Scores
avg_scores = pd.DataFrame(school_data_complete.groupby("school_name")
                          ["math_score", "reading_score"].mean())
avg_scores.reset_index(inplace=True)
avg_scores.columns = ["School Name", "Average Math Score", "Average Reading Score"]
# avg_scores
```

```python
# Merge the two created data frames on the School Name
merged_df = pd.merge(schools, avg_scores, on="School Name")
# merged_df
```

```python
# Calculate number of students with a passing math score (70 or greater)
# Created data frame in District data for only students achieving greater than 69 in math
math_gp = pd.DataFrame(pass_math.groupby("school_name")
                       ["math_score"].count())
math_gp.reset_index(inplace=True)
math_gp.columns = ["School Name", "Passing Math"]
# math_gp
```

```python
# Calculate the number of students with a passing reading score (70 or greater)
# Created data frame in District data for only students achieving greater than 69 in reading
read_gp = pd.DataFrame(pass_read.groupby("school_name")
                       ["reading_score"].count())
read_gp.reset_index(inplace=True)
read_gp.columns = ["School Name", "Passing Reading"]
# read_gp
```

```python
# Merge the two created data frames on the School Name
math_read = pd.merge(math_gp, read_gp, on="School Name")
merged_df = pd.merge(merged_df, math_read, on="School Name")
# merged_df
```

```python
# Calculate percentage of students passing math and add column
merged_df["% Passing Math"] = merged_df["Passing Math"]/merged_df["Total Students"] * 100

# Calculate percentage of students passing reading and add column
merged_df["% Passing Reading"] = merged_df["Passing Reading"]/merged_df["Total Students"] * 100

# Overall Passing Rate (Average of the above two), add column
merged_df["Overall Passing Rate"] = (merged_df["% Passing Math"] + merged_df["% Passing Reading"])/2
merged_df.drop(["Passing Math","Passing Reading"],axis=1, inplace=True)

# Calculate Per Student Budget, add column
merged_df["Budget per Student"] = merged_df["Total Budget"]/merged_df["Total Students"]

# merged_df
```

```python
# For reference only
# merged_df.drop(["col1","col2"],axis=1, inplace=True)
```

```python
# Format and set layout for overview table
merged_df = merged_df.rename(columns={"Type":"School Type",
                          "Total Budget":"Total School Budget",
                          "Budget per Student": "Per Student Budget",
                          "Overall Passing Rate":"% Overall Passing Rate"
                            })
merged_df = merged_df[["School Name",
                       "School Type",
                       "Total Students", 
                       "Total School Budget", 
                       "Per Student Budget",
                       "Average Math Score",
                       "Average Reading Score",
                       "% Passing Math",
                       "% Passing Reading",
                       "% Overall Passing Rate"
                       ]]
merged_df.set_index("School Name", inplace=True)

# Optional: give the displayed data cleaner formatting
    # Total budget: Format with dollar sign and comma
merged_df["Total School Budget"] = merged_df["Total School Budget"].astype(float).map("${:,.2f}".format)
    # Per Student budget: Format with dollar sign and .00
merged_df["Per Student Budget"] = merged_df["Per Student Budget"].astype(float).map("${:.2f}".format)
    # Total students: Format with comma
merged_df["Total Students"] = merged_df["Total Students"].map("{:,}".format)
    # Average Math and Reading; Overall Rate; Math and Reading passing: Format to 6 decimal places
merged_df["Average Math Score"] = merged_df["Average Math Score"].map("{:.6f}".format)
merged_df["Average Reading Score"] = merged_df["Average Reading Score"].map("{:.6f}".format)
merged_df["% Passing Math"] = merged_df["% Passing Math"].map("{:.6f}".format)
merged_df["% Passing Reading"] = merged_df["% Passing Reading"].map("{:.6f}".format)
merged_df["% Overall Passing Rate"] = merged_df["% Overall Passing Rate"].map("{:.6f}".format)
merged_df

merged_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
merged_df.loc[:,"Per Student Budget"] = pd.to_numeric(merged_df.loc[:,"Per Student Budget"].str.replace('$',''))
# merged_df.dtypes

merged_df.reset_index()
# merged_df

merged_df["Spending Ranges (Per Student)"] = pd.cut(merged_df["Per Student Budget"], spending_bins, 
                                              labels=group_names)
# merged_df
```

```python
# Create a table that breaks down school performances based on average Spending Ranges (Per Student).
spend_df = merged_df.groupby("Spending Ranges (Per Student)") ["Average Math Score",
                                                                 "Average Reading Score",
                                                                 "% Passing Math",
                                                                 "% Passing Reading",
                                                                 "% Overall Passing Rate"].max()
spend_df
```
